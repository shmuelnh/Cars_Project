### List of all cars and links

```python
car_list = list() 
pages = ['1','2','3'] #there is 3 pages
for i in pages:
    url = "https://www.ad.co.il/car?keyword=%D7%90%D7%90%D7%95%D7%93%D7%99&pageindex="+i
    html = requests.get(url)
    soup = BeautifulSoup(html.content,'html.parser')
    cars = soup.find_all('div',class_="card-body p-md-3")
    for car in cars:
        link = "https://www.ad.co.il" + car.find('a').get('href')
        name = car.find('a').get_text()
        car_list.append((name,link)) 
        
```

### Function that returns car data

```python
def get_car_info(car_link):
    car_dict = {}
    try:
        response = requests.get(car_link)
        if not response.status_code == 200:
            return car_dict
        car_soup = BeautifulSoup(response.content,'html.parser')
        try:
            car_dict['Model']=car_soup.find('div',class_="d-flex justify-content-between").find_all('h2')[0].get_text()
        except:
            car_dict['Model']=None
        try:
            car_dict['Price']=car_soup.find('div',class_="d-flex justify-content-between").find_all('h2')[1].get_text()
        except:
            car_dict['Price']=None
        table = car_soup.find('table',class_="table table-sm mb-4") 
        for row in table.find_all('tr'):
            two_columns = row.find_all('td')
            column_name = two_columns[0].text.strip()
            value = two_columns[1].text.strip()
            car_dict[column_name] = value
        try:
            car_dict['Image_num'] = len(car_soup.find('div',class_="col-12 d-flex mt-3 justify-content-center flex-wrap").find_all('div',class_="justify-content-center px-1"))
        except:
            car_dict['Image_num']=None
        try:
            car_dict['Description']=car_soup.find('p',class_='text-word-break').get_text().strip()
        except:
            car_dict['Description']=None
        try:
            divs = car_soup.find_all('div', class_='px-3')
            dates = [re.search(r'\d{2}/\d{2}/\d{4}', div.get_text()).group() for div in divs if re.search(r'\d{2}/\d{2}/\d{4}', div.get_text())]
            if len(dates) > 0:
                car_dict['Cre_date'] = dates[0]
            else:
                car_dict['Cre_date']= None
        except:
            car_dict['Cre_date']= None
        try:
            if len(dates) > 1:
                car_dict['Repub_date'] = dates[1]
            else:
                car_dict['Repub_date']= None
        except:
            car_dict['Repub_date']= None

        return car_dict
    except:
        return car_dict
```

#### Makeing a dataframe

#### Rename and reorder columns

```python
Audi_df['Manufactor']='אאודי'
Audi_df['Model']=Audi_df['Model'].str.replace('אאודי',"")
Audi_df
```

```python
Audi_df=Audi_df.rename(columns={'Image_num':'Pic_num','שנה':'Year','יד':'Hand','ת. הילוכים':'Gear','נפח':'capacity_Engine','סוג מנוע':'Engine_type','ק"מ':'Km','צבע':'Color','בעלות קודמת':'Prev_ownership','בעלות נוכחית':'Curr_ownership','אזור':'Area','עיר':'City','טסט עד':'Test'})
Audi_df
```

```python
Audi_df.columns
```

```python
new_order = ['Manufactor','Year','Model','Hand','Gear','capacity_Engine','Engine_type','Prev_ownership','Curr_ownership','Area','City','Price','Pic_num','Cre_date','Repub_date','Description','Color','Km','Test']
```

```python
Audi_df = Audi_df[new_order]
```

#### Changing type

```python
Audi_df.info()
```

```python
from datetime import datetime
Audi_df['Test'] = pd.to_datetime(Audi_df['Test'], errors='coerce')
Audi_df['Test'] = (Audi_df['Test'] - pd.to_datetime(datetime.today().strftime('%Y-%m-%d'))).dt.days
```

```python
Audi_df['Km'] = Audi_df['Km'].str.replace(r'[^\d]', '', regex=True)
Audi_df['Km'] = pd.to_numeric(Audi_df['Km'], errors='coerce').astype('Int64')
```

```python
Audi_df['capacity_Engine'] = Audi_df['capacity_Engine'].str.replace(r'[^\d]', '', regex=True)
Audi_df['capacity_Engine'] = pd.to_numeric(Audi_df['capacity_Engine'], errors='coerce').astype('Int64')
```

```python
Audi_df['Test'] = pd.to_numeric(Audi_df['Test'], errors='coerce').astype('Int64')
```

```python
Audi_df['Year'] = Audi_df['Year'].astype(int)
Audi_df['Cre_date'] = pd.to_datetime(Audi_df['Cre_date'], format='%d/%m/%Y')
Audi_df['Repub_date'] = pd.to_datetime(Audi_df['Repub_date'], format='%d/%m/%Y')
Audi_df['Prev_ownership'] = Audi_df['Prev_ownership'].astype('category') 
Audi_df['Curr_ownership'] = Audi_df['Curr_ownership'].astype('category')
Audi_df['Manufactor'] = Audi_df['Manufactor'].astype(str)
Audi_df['Model'] = Audi_df['Model'].astype(str)
Audi_df['Area'] = Audi_df['Area'].astype(str)
Audi_df['City'] = Audi_df['City'].astype(str)
Audi_df['Description'] = Audi_df['Description'].astype(str)
Audi_df['Hand'] = Audi_df['Hand'].astype(int)
Audi_df['Gear'] = Audi_df['Gear'].astype('category')
Audi_df['Engine_type'] = Audi_df['Engine_type'].astype('category')
Audi_df['Color'] = Audi_df['Color'].astype(str)
```

```python
Audi_df['Pic_num'] = pd.to_numeric(Audi_df['Pic_num'], errors='coerce').astype('Int64')
```

```python
Audi_df['Price'] = Audi_df['Price'].str.replace(r'[^\d]', '', regex=True)
Audi_df['Price'] = pd.to_numeric(Audi_df['Price'], errors='coerce').astype('float')
```
